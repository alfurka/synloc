# Linear Regression

```python
import statsmodels.api as sm
from statsmodels.iolib.summary2 import summary_model, summary_col
```

```python
y_list = ['lannwage']
x_list = ['constant','age','educ','hrsly']
```

```python
def prep_data(df):
    dataset = df.copy()
    dataset['lannwage'] = np.log(1+dataset['annwage'])
    dataset['constant'] = 1
    dataset = dataset[y_list + x_list]
    return dataset
```

```python
ols_data = prep_data(df)
model_main = sm.OLS(ols_data[y_list], ols_data[x_list], hasconst= True).fit()
ols_data = prep_data(dfSyn) 
model_shuffle = sm.OLS(ols_data[y_list], ols_data[x_list], hasconst= True).fit()
ols_data = prep_data(dfSyn1) 
model_LocCov = sm.OLS(ols_data[y_list], ols_data[x_list], hasconst= True).fit()
ols_data = prep_data(dfSyn2) 
model_copula = sm.OLS(ols_data[y_list], ols_data[x_list], hasconst= True).fit()
ols_data = prep_data(dfSyn3) 
model_ctgan = sm.OLS(ols_data[y_list], ols_data[x_list], hasconst= True).fit()
ols_data = prep_data(dfSyn4) 
model_copulaGan = sm.OLS(ols_data[y_list], ols_data[x_list], hasconst= True).fit()
ols_data = prep_data(dfSyn5) 
model_synthpop = sm.OLS(ols_data[y_list], ols_data[x_list], hasconst= True).fit()
```

```python
result_table = summary_col([model_main, 
             model_shuffle, 
             model_LocCov, 
             model_copula, 
             model_ctgan, 
             model_copulaGan, 
             model_synthpop ],
           stars = True,
            float_format='%.2f',
           model_names= ['Original Sample',
                        'LCE',
                        'LCE without Shuffling',
                        'Gaussian Copula',
                        'CTGAN',
                        'CopulaGAN',
                        'synthpop'])
```
